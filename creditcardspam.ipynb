importing the dependencies mainly


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#now we are updating the csv file
credit_card=pd.read_csv('/content/creditcard.csv')
#now we are printing  the first 5 rows of the dataset
credit_card.head()
```

now we are going to learn abut the missing values in the data set


```python
credit_card.isnull().sum()
#distribution of the fraudetection and legimit transcationnnnnnnnn
credit_card['Class'].value_counts()
```

this dataset is highly unblanceddd

```python
#0-means good transcation
#1-means fraud transcation
#seperating the good and bad transcations
legit=credit_card[credit_card.Class==0]
fraud=credit_card[credit_card.Class==1]
print(legit.shape)
print(fraud.shape)

#now we are chcking the statistical dataaaaaaaaaaaa
legit.Amount.describe()
fraud.Amount.describe()
#compare the values of the for each transcationnn
credit_card.groupby('Class').mean()
```

```python
#concatenating two dataframes
legit_sample=legit.sample(n=492)
new_data=pd.concat([legit_sample,fraud],axis=0)
new_data.head()
```

```python
#now we are new_dataset['Class'].value_counts()
new_data['Class'].value_counts()
new_data.groupby('Class').mean()
```

now we are splitting the data in to the features and targetss


```python
X=new_data.drop(columns='Class',axis=1)
Y=new_data['Class']
print(X)
#now here we are not having the any target column only features are present in the x
```

```python
#now split the data in to train_data and test data
#X is the featuressssss Y is the classsssssssss
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
print(X.shape,X_train.shape,X_test.shape)
```

now we are perfrming the mdel training


```python
#using the logistic regresssion we are performing this ttassssk
#now we have to train the data with logisticregresssin model
model=LogisticRegression()
model.fit(X_train,Y_train)
#accuracy socre on the training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
print('accuracy on training data:',training_data_accuracy)
#acuuracy on the test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print('accuracy on test data:',test_data_accuracy)
```
